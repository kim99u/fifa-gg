```python
def dictsetter():
    temp_dict = {}
    keys = ['pay', 'overall', 'position', 'age', 'height', 'weight', 'bodytype', 'skill', 'leftfoot', 'rightfoot', \
            'reputation', 'nation', 'traits', 'ovr_pace', 'ovr_shoot', 'ovr_pass', 'ovr_dribble', 'ovr_defence', 'ovr_physical', \
            'sprint', 'accel', 'finish', 'shotpower', 'longshot', 'positioning', 'volley', 'penalty', 'shortpass', \
            'vision', 'cross', 'longpass', 'freekick', 'curve', 'dribble', 'control', 'agility', 'balance', 'reaction', \
            'marking', 'tackle', 'intercept', 'header', 'sliding', 'strength', 'stamina', 'aggression', 'jump', \
            'composure', 'gk_diving', 'gk_handle', 'gk_kick', 'gk_reflex', 'gk_positioning', 'teamcolor', 'clubname', \
            'clubimg', 'mugshot']
    for key in keys:
        temp_dict[key] = [] 
    return temp_dict
```

```python
def crawler (id_df, temp_dict):
    for x in tqdm(id_df['id']):
        url = f'https://fifaonline4.nexon.com/DataCenter/PlayerInfo?spid={x}&n1Strong=0'
        ranheaders = {'User-Agent': generate_user_agent(os='win', device_type='desktop')}
        html = requests.get(url, headers=ranheaders).text
        soup = BS(html, 'html.parser')
        body = soup.find('body')

        info_wrap = body.find('div', class_='info_wrap')
        middle = body.find('div', class_='content_middle').find('ul')
        middle_values = middle.findAll('div', class_=re.compile('value +'))
        bottom = body.find('div', class_='content_bottom')
        bottom_wrap = bottom.findAll('ul', class_=re.compile('data_wrap+'))
        traits = info_wrap.find('div', class_='skill_wrap').findAll('span')
        temptraits = ''
        for trait in traits:
            temptraits += trait.text.strip() + ', '

        temp_dict['pay'].append(info_wrap.find('div', class_='pay_side').text.strip())
        temp_dict['overall'].append(body.find('div', class_='ovr value').text.strip())
        temp_dict['position'].append(body.find('div', class_='position').text.strip())
        temp_dict['age'].append(info_wrap.find('span', class_='etc birth').text.strip())
        temp_dict['height'].append(info_wrap.find('span', class_='etc height').text.strip()[:3])
        temp_dict['weight'].append(info_wrap.find('span', class_='etc weight').text.strip()[:2])
        temp_dict['bodytype'].append(info_wrap.find('span', class_='etc physical').text.strip())
        temp_dict['skill'].append(len(info_wrap.find('span', class_='etc skill').find('span').text.strip()))
        temp_dict['leftfoot'].append(info_wrap.find('span', class_='etc foot').text.strip()[1])
        temp_dict['rightfoot'].append(info_wrap.find('span', class_='etc foot').text.strip()[-1])
        temp_dict['reputation'].append(info_wrap.find('span', class_='etc season').text.strip())
        temp_dict['nation'].append(info_wrap.find('div', class_='etc nation').find('span', class_='txt').text.strip().replace(', 국가대표', ''))
        temp_dict['traits'].append(temptraits.strip(', '))
        temp_dict['ovr_pace'].append(middle_values[0].text.strip())
        temp_dict['ovr_shoot'].append(middle_values[1].text.strip())
        temp_dict['ovr_pass'].append(middle_values[2].text.strip())
        temp_dict['ovr_dribble'].append(middle_values[3].text.strip())
        temp_dict['ovr_defence'].append(middle_values[4].text.strip())
        temp_dict['ovr_physical'].append(middle_values[5].text.strip())
        temp_dict['sprint'].append(bottom_wrap[0].findAll('div', class_=re.compile('value+'))[0].text.strip())
        temp_dict['accel'].append(bottom_wrap[0].findAll('div', class_=re.compile('value+'))[1].text.strip())
        temp_dict['finish'].append(bottom_wrap[0].findAll('div', class_=re.compile('value+'))[2].text.strip())
        temp_dict['shotpower'].append(bottom_wrap[0].findAll('div', class_=re.compile('value+'))[3].text.strip())
        temp_dict['longshot'].append(bottom_wrap[0].findAll('div', class_=re.compile('value+'))[4].text.strip())
        temp_dict['positioning'].append(bottom_wrap[0].findAll('div', class_=re.compile('value+'))[5].text.strip())
        temp_dict['volley'].append(bottom_wrap[0].findAll('div', class_=re.compile('value+'))[6].text.strip())
        temp_dict['penalty'].append(bottom_wrap[1].findAll('div', class_=re.compile('value+'))[0].text.strip())
        temp_dict['shortpass'].append(bottom_wrap[1].findAll('div', class_=re.compile('value+'))[1].text.strip())
        temp_dict['vision'].append(bottom_wrap[1].findAll('div', class_=re.compile('value+'))[2].text.strip())
        temp_dict['cross'].append(bottom_wrap[1].findAll('div', class_=re.compile('value+'))[3].text.strip())
        temp_dict['longpass'].append(bottom_wrap[1].findAll('div', class_=re.compile('value+'))[4].text.strip())
        temp_dict['freekick'].append(bottom_wrap[1].findAll('div', class_=re.compile('value+'))[5].text.strip())
        temp_dict['curve'].append(bottom_wrap[1].findAll('div', class_=re.compile('value+'))[6].text.strip())
        temp_dict['dribble'].append(bottom_wrap[2].findAll('div', class_=re.compile('value+'))[0].text.strip())
        temp_dict['control'].append(bottom_wrap[2].findAll('div', class_=re.compile('value+'))[1].text.strip())
        temp_dict['agility'].append(bottom_wrap[2].findAll('div', class_=re.compile('value+'))[2].text.strip())
        temp_dict['balance'].append(bottom_wrap[2].findAll('div', class_=re.compile('value+'))[3].text.strip())
        temp_dict['reaction'].append(bottom_wrap[2].findAll('div', class_=re.compile('value+'))[4].text.strip())
        temp_dict['marking'].append(bottom_wrap[2].findAll('div', class_=re.compile('value+'))[5].text.strip())
        temp_dict['tackle'].append(bottom_wrap[2].findAll('div', class_=re.compile('value+'))[6].text.strip())
        temp_dict['intercept'].append(bottom_wrap[3].findAll('div', class_=re.compile('value+'))[0].text.strip())
        temp_dict['header'].append(bottom_wrap[3].findAll('div', class_=re.compile('value+'))[1].text.strip())
        temp_dict['sliding'].append(bottom_wrap[3].findAll('div', class_=re.compile('value+'))[2].text.strip())
        temp_dict['strength'].append(bottom_wrap[3].findAll('div', class_=re.compile('value+'))[3].text.strip())
        temp_dict['stamina'].append(bottom_wrap[3].findAll('div', class_=re.compile('value+'))[4].text.strip())
        temp_dict['aggression'].append(bottom_wrap[3].findAll('div', class_=re.compile('value+'))[5].text.strip())
        temp_dict['jump'].append(bottom_wrap[3].findAll('div', class_=re.compile('value+'))[6].text.strip())
        temp_dict['composure'].append(bottom_wrap[4].findAll('div', class_=re.compile('value+'))[0].text.strip())
        temp_dict['gk_diving'].append(bottom_wrap[4].findAll('div', class_=re.compile('value+'))[1].text.strip())
        temp_dict['gk_handle'].append(bottom_wrap[4].findAll('div', class_=re.compile('value+'))[2].text.strip())
        temp_dict['gk_kick'].append(bottom_wrap[4].findAll('div', class_=re.compile('value+'))[3].text.strip())
        temp_dict['gk_reflex'].append(bottom_wrap[4].findAll('div', class_=re.compile('value+'))[4].text.strip())
        temp_dict['gk_positioning'].append(bottom_wrap[4].findAll('div', class_=re.compile('value+'))[5].text.strip())
        club_target = body.find('div', class_='etc team')
        club = club_target.find('span', class_='txt')
        if (club == None):
            temp_dict['clubname'].append(None)
            temp_dict['clubimg'].append(None)
        else:
            club_txt = club.text.strip()
            club_img = club_target.find('img')['src'].strip()
            if (len(club_txt) == 0):
                temp_dict['clubname'].append(None)
            else:
                temp_dict['clubname'].append(club_txt)
            temp_dict['clubimg'].append(club_img)
        temp_dict['mugshot'].append(body.find('div', class_='img').find('img')['src'][:-16])
        team_list = body.findAll('div', class_='td club')
        team_str = ''
        for team in team_list:
            team_str += (team.text.strip() + ', ')
        temp_dict['teamcolor'].append(team_str.rstrip(', '))

    temp_df = pd.DataFrame(data = temp_dict)
    return temp_df
```

```python
crawled_df = crawler(id_df)
```

```python
info_df = pd.concat([id_df, crawled_df], axis=1)
info_df
```
